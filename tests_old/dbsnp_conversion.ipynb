```python
from handygenome.common import ChromDict

FASTA_PATH_HG19 = "/home/users/data/01_reference/human_g1k_v37/human_g1k_v37.fasta"
FASTA_PATH_HG38 = "/home/users/data/01_reference/human_g1k_v38/Homo_sapiens_assembly38.fasta"

FASTA_HG19 = pysam.FastaFile(FASTA_PATH_HG19)
FASTA_HG38 = pysam.FastaFile(FASTA_PATH_HG38)
CHROMDICT_HG19 = ChromDict(fasta_path=FASTA_PATH_HG19)
CHROMDICT_HG38 = ChromDict(fasta_path=FASTA_PATH_HG38)
```

```python
from handygenome.dbconv import dbsnp as libdbsnpconv
```

```python
original_vcf_path = '/home/users/pjh/References/dbSNP37/original_files_220218/GCF_000001405.25.gz'
outfile_path = '/home/users/pjh/scripts/python_genome_packages/data/popfreq/dbSNP_b155_GRCh37.p13.vcf.gz'
refver = 'GRCh37'
chromdict = CHROMDICT_HG19
fasta = FASTA_HG19
```

```python
chrname_converter = libdbsnpconv.get_chrname_converter(refver)
```

```python
import tempfile
tmpdir = tempfile.mkdtemp(prefix='dbsnp_converter', dir=os.path.dirname(outfile_path))
```

```python
chromlist, tmpfile_paths = libdbsnpconv.get_tmpfile_paths(tmpdir, original_vcf_path, chromdict, chrname_converter)
```

```python
popnames = libdbsnpconv.preprocess(original_vcf_path, chromlist, tmpfile_paths, fasta, chromdict, logging_lineno=5_000_000)
```

```python
libdbsnpconv.write_final_outfile(popnames, chromdict, tmpfile_paths, outfile_path, chromlist, logging_lineno=5_000_000)
```
